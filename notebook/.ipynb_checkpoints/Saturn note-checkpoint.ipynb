```python
y = df['Class']
X = df.drop(columns='Class', axis=1)
```

```python
X.shape, y.shape
```

```python
smk = SMOTETomek(random_state = 42)
X_ov_smk, y_ov_smk = smk.fit_sample(X,y)
```

```python
y_temp = pd.DataFrame(data=y_ov_smk, columns=['Class'])
ax = sns.countplot(x="Class", data=y_temp)
plt.title("Transaction Class Distribution")
plt.xlabel("Class")
plt.ylabel("Frequency");
```

### Data splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_ov_smk, y_ov_smk, test_size=0.4, random_state=42)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
clf_model = RandomForestClassifier()
clf_model.fit(X_train, y_train)
```

```python
def show_model_report(x,y,model):
    y_pred = model.predict(x)

    print('Classification report:')
    print(classification_report(y, y_pred))
    print("-----------------------------------\n")
    print('Confusion Metrix: ')
    print(confusion_matrix(y, y_pred))
    print("-----------------------------------\n")
    print('Accuracy Score')
    print(roc_auc_score(y, y_pred))
    print("-----------------------------------\n")
    
    try:
        y_pred = model.predict_proba(x)[:,1]
        FPR, TPR, threshold = roc_curve(y, y_pred)
        print("ROC AUC Score")
        print(roc_auc_score(y, y_pred))

        #Plot ROC curve
        plt.title('ROC Curve')
        plt.plot(FPR, TPR)
        plt.plot([0,1], ls ='--')
        plt.plot([0,0], [1,0], c='.7'), plt.plot([1,1], c='.7')
        plt.ylabel('True Positive Rate')
        plt.xlabel('False Positive Rate')
        plt.show()
    except Exception as ex:
        pass
```

```python
print("Train data report")
print("---------------------------------------")
show_model_report(X_train,y_train,clf_model)
print('\n\n')
print("Test data report")
print("---------------------------------------")
show_model_report(X_test,y_test,clf_model)
```
